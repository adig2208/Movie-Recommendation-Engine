#Demographic Filtering

Weighted Rating Formula = [(v/(v+m)).R] + [(m/(v+m)).C]

where,

v is the number of votes for the movie;

m is the minimum votes required to be listed in the chart;

R is the average rating of the movie; And

C is the mean vote across the whole report

```python
C = merged_df['vote_average'].mean()
m = merged_df['vote_count'].quantile(0.9)
C,m
```

```python
demo_movies = merged_df.copy().loc[merged_df['vote_count'] >= m]
demo_movies
```

```python
def weighted_rating(val,C=C,m=m):
  v = val['vote_count']
  R = val['vote_average']
  return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
demo_movies['score'] = demo_movies.apply(weighted_rating, axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
merged_df['overview'] = merged_df['overview'].fillna('')
matrix = vectorizer.fit_transform(merged_df['overview'])
matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(matrix)
```

```python
def recommend_movies(title, similarity_matrix=similarity_matrix, titles=merged_df['title'], indices=pd.Series(merged_df.index, index=merged_df['title'])):
    idx = indices[title]
    sim_scores = list(enumerate(similarity_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```

```python
recommend_movies('Interstellar')
```

```python
def plot_movie_similarity(title, similarity_matrix=similarity_matrix, titles=merged_df['title'], indices=pd.Series(merged_df.index, index=merged_df['title'])):
    idx = indices[title]
    sim_scores = list(enumerate(similarity_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]

    sim_movies = titles.iloc[movie_indices]
    sim_scores = [i[1] for i in sim_scores]

    plt.figure(figsize=(12,6))
    sns.barplot(x=sim_scores, y=sim_movies, palette='viridis')
    plt.title(f'Movies Similar to {title}')
    plt.xlabel('Similarity Score')
    plt.ylabel('Movies')
    plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertModel, BertTokenizer
import torch
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
def get_bert_embeddings(texts, tokenizer, model, device='cpu'):
    model.to(device)
    embeddings = []

    for text in texts:
        inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512).to(device)
        with torch.no_grad():
            outputs = model(**inputs)

        emb = outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
        embeddings.append(emb)

    return embeddings
```

```python
subset = merged_df['overview'].fillna("")

embeddings = get_bert_embeddings(subset, tokenizer, model)

print(f"Generated embeddings shape: {len(embeddings)} x {len(embeddings[0])}")
```

```python
import numpy as np
embeddings_matrix = np.array(embeddings)

similarity_matrix = cosine_similarity(embeddings_matrix)

def recommend_movies_bert(title, similarity_matrix, titles):
    indices = pd.Series(range(len(titles)), index=titles)

    try:
        idx = indices[title]
        sim_scores = list(enumerate(similarity_matrix[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        movie_indices = [i[0] for i in sim_scores[1:11]]
        return titles.iloc[movie_indices]
    except IndexError:
        print(f"Title '{title}' not found in the dataset.")
        return []


recommend_movies_bert('Interstellar', similarity_matrix, merged_df['title'])
```

```python
def create_prompt(title, genres, overview):
    if isinstance(genres, str):
        genres_list = eval(genres)
    else:
        genres_list = genres
    if isinstance(genres_list, list) and all(isinstance(item, dict) for item in genres_list):
        genres_str = ', '.join([genre['name'] for genre in genres_list])
    else:
        genres_str = ""

    prompt = f"Write a review for the movie '{title}', a {genres_str} film. Synopsis: {overview}"
    return prompt
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load the tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Set the padding token
tokenizer.pad_token = tokenizer.eos_token

# Define the function to generate a review
def generate_review(prompt):
    # Encode the prompt
    inputs = tokenizer.encode(prompt, return_tensors='pt', padding=True, truncation=True, max_length=512)
    # Generate the sequence
    outputs = model.generate(inputs, max_length=100, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
    # Decode and return the generated text
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Example usage
for index, row in merged_df.head(5).iterrows():
    prompt = create_prompt(row['title'], row['genre_list'], row['overview'])
    review = generate_review(prompt)
    print(f"Generated Review for {row['title']}: {review}\n")
```

```python
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split

reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(ratings_small_df[['userId', 'movieId', 'rating']], reader)
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
from surprise import SVD, accuracy
from surprise.model_selection import cross_validate

algo = SVD()
algo.fit(trainset)
predictions = algo.test(testset)

accuracy.rmse(predictions)
```

```python
from collections import defaultdict

def get_top_n(predictions, n=10):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
top_n = get_top_n(predictions, n=5)
```

```python
def get_recommendations(user_id, model, data, n=10):
    movie_ids = movies_with_ratings_df['movieId'].unique()
    user_rated_movies = movies_with_ratings_df[movies_with_ratings_df['userId'] == user_id]['movieId']
    unrated_movies = [movie for movie in movie_ids if movie not in user_rated_movies.values]

    predictions = [model.predict(user_id, movie_id) for movie_id in unrated_movies]
    predictions.sort(key=lambda x: x.est, reverse=True)
    top_n_movie_ids = [pred.iid for pred in predictions[:n]]

    top_n_titles = movies_metadata_df[movies_metadata_df['id'].isin(top_n_movie_ids)]['title']
    return top_n_titles.tolist()
```

```python
recommended_movie_ids = get_recommendations(user_id=1, model=algo, data=data, n=5)
print(recommended_movie_ids)
```
